```python
text="My most recent trek to the Himalayas was a test of both stamina and spirit. I will never forget the moment the sun finally broke through the heavy morning mist at the summit, turning the jagged, snow-capped peaks into a shimmering wall of gold. Standing there with my friends, gasping for the thin, cold air, I felt a profound sense of peace that made every grueling mile of the ascent worth it."
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
tokens = word_tokenize(text)
print(tokens)
```

```python
stemmer = PorterStemmer()
stem = [stemmer.stem(word) for word in tokens]
print(stem)
```

```python
lemmaizer = WordNetLemmatizer()
lemma = [lemmaizer.lemmatize(word) for word in tokens]
print(lemma)
```

```python
import nltk
nltk.download('averaged_perceptron_tagger_eng')
```

```python
pos_tags = pos_tag(tokens)
print(pos_tags)
```

```python
for word, tag in pos_tags:
    print(f"{word} → {tag}")
```

```python
sia = SentimentIntensityAnalyzer()
sentiment = sia.polarity_scores(text)
print(sentiment)
```
