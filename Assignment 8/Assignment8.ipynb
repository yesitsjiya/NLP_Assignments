```python
from nltk.corpus import wordnet as wn
from nltk.wsd import lesk
from nltk.tokenize import word_tokenize
```

```python
sentence = "He went to the bank to deposit his money."
word = "bank"

sense = lesk(word_tokenize(sentence), word)
print("Best sense:", sense)
print("Definition:", sense.definition())
```

```python
sentence2 = "The fisherman sat on the bank of the river."
sense2 = lesk(word_tokenize(sentence2), "bank")

print("Sentence:", sentence2)
print("Best sense:", sense2)
print("Definition:", sense2.definition())
```

```python
def disambiguate_sentence(sentence, target_word):
    tokens = word_tokenize(sentence)
    sense = lesk(tokens, target_word)
    
    if sense:
        return {
            "word": target_word,
            "sense": sense.name(),
            "definition": sense.definition(),
            "examples": sense.examples()
        }
    else:
        return None
```

```python
sentences = [
    "She made a deposit at the bank.",
    "They sat on the bank of the lake.",
    "He went to the bat store.",
    "The bat flew out of the cave."
]

targets = ["bank", "bank", "bat", "bat"]

for sent, word in zip(sentences, targets):
    result = disambiguate_sentence(sent, word)
    print(f"\nSentence: {sent}")
    print("Word:", word)
    print("Sense:", result["sense"])
    print("Definition:", result["definition"])
```
